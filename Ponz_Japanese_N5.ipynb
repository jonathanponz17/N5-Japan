##Grammar 1: N1 **は** N２ **です**

###Part 1: Particle **は**

*Note:* The particle **は** is read **わ**

The particle **は** indicates that the words before it is the topic of the sentence. You select a noun you want to talk about, add は to show that it is the topic and give a statement about the topic.

###Part 2:**です**

Nouns used with **です** work as predicates.

**です** indicates judgement or assertion

**です** also conveys that the speaker is being polite towards the listener

###Examples:

1. わたし **は**  マイク ミラー**です**
    
    i  am  mike miller 

2. わたし **は**  フィリピン じん **です**

    i am filipino

3. わたし  **は**  かいしゃいん **です**

    I am an office worker

4. ケビン **は**  にじゅうごさい **です**

    kevin is 25 years old 

## Grammar 2: N1 **は** N2  **じゃ ありません**

###Part 1: **じゃ ありません**

**じゃ ありません** is the negative form of **です**. It is the form used in daily conversation. 

For a formal speech or writting, **では ありません** is used instead.
*Note:* **は** in **では** is read **わ**.


###Examples:

1. サントスさん**は**  がくせい**じゃ ありません**

    Mr.Santos is not a student.
          Mr. Santos "wa" student "ja arimesen"

2. わたし  **は**  にほんじん  **じゃ ありません**

    I'm not a Japanese
          I "wa" Japanse "ja arimasen"

3. やまだ  **は**  エンジニア  **じゃ ありません**

    Yamada is not an engineer 
          Yamada "wa" engineer "ja arimasen"

##Grammar 3: S   **か**

Particle **か**

The particle **か** is used to express the speaker’s doubt, question, uncertainty, etc. A question is formed by simply adding **か** to the end of the sentence. A question ends with a rising intonation.

###Part 1: Questions asking whether a statement is correct or not

As mentioned above, a sentence becomes a question when **か** is added to the end. The word order does not change. The question thus made asks whether a statement is correct or not. Depending on whether you agree with the statement or not, your answer to such a question begins with **はい** or **いいえ**.

####Examples:

1. Question: あなた は せんせい です**か**？ 

    Are you a teacher?

  *Ans#1*: 
  
  はい わたし は せんせい です
    
    Yes i am a teacher

  *Ans#2:* 
  
  いいえ わたし は せんせい じゃ ありません
    
    No I'm not a teacher

  *Follow-up Ans#2:* 
  
  わたし は ぎんこういん です

    I am a bank employee

2. **Question:** あなた は にほんじん です **か**？

    Are you a Japanese?

    *Ans#1*: 
  
  はい わたし は にほんじん です
    
    yes i am a japanese

  *Ans#2:* 
  
  いいえ わたし は にほんじん じゃ ありません
    
    no i am not a japanese 

  *Follow-up Ans#2:* 
  
  わたし は  フィリピン じん    です

    i am  filipino

##Grammar 4: N  particle **も**

**も** is added after a topic instead of は when the statement about the topic is the same as the previous topic. 
**Also**

###Examples:

1. ミラーさんは かいしゃいんです. グプタさん**も** かいしゃいんです.

  Mr.Miller is a company employee. Mr.Gupta is **also** a company employee.

2. わたし は がくせい です. Mikey **も** がくせい です.

  I am a student. Mikey is **also** a student.

##Grammar 5: N1 **の** N2

**の** is used to connect two nouns. N1 modifies N2 . In Lesson 1, N1 is an organization or some kind of group to which N2 belongs

###Examples:

1. ミラーさんは IMC **の** しゃいんです.

  Mr.Miller is an IMC employee.

        Mr. Miller "wa" IMC "no" employee "desu"

2. わたし は しんじる  **の** がくせい です.

  I am a student of shinjiru.

        I "wa" shinjiru "no" student "desu".

3. わたし は  みつびしい **の** エンジニア です.

  I am an engineer of mitsubishi

        I "wa" Mitsubishi "no" engineer "desu".

##Grammar 6: ~**さん**

When referring directly to the listener’s name. The listener’s family name followed by さん is usually used.

###Examples:

#Lesson 2: Grammar

##Part 1: **これ** /  **それ** / **あれ**

**これ**, **それ** and **あれ** are demonstratives.

They work as nouns. 

**これ** refers to a thing  near the speaker. 

  *Sentence Pattern:* **これ** + **は** + N2 + **です**

**それ** refers to a thing near the listener.  

  *Sentence Pattern:* **それ** + **は** + N2 + **です**

**あれ** refers to a thing  far from the speaker and listener.

  *Sentence Pattern:* **あれ** + **は** + N2 + **です**



###Examples: Declarative

1. **これ** は ほん です

  **This is** a book

2. **それ** は じしょ です

  **That** is a dictionary.

3. **あれ** は かばん です.

  **That** is a bag over there.

###Examples: Confirmation Question ( か )?

1. **これ** は **えんぴつ** です か？

  Is **this** a *pencil*?

2. **それ** は **かぎ** です か？

  Is **that** a *key*?

3. **あれ** は **てちょう**  ですか？

  Is **that** a *pocket notebook*?

###Examples: Interrogative Question  (なん です か)？

1. **これ** は なん です か？

  What is **this**?

2. **それ** は なん です か？

  What is **that**?

3. **あれ** は なん です か？

  What is **that thing over there**?

##Part 2: **この N** / **その N** / **あの N**

**この**,  **その**,  and **あの** modify nouns.

“**この N**” refers to a thing  or a person near the speaker.

“**その N**” refers to a thing  or a person near the listener.

“**あの N**” refers to a thing  or a person far from both are speaker and the listener.



###Examples: Simple

1. **この** くるま.

  **This** car.

2. **その** かばん.

  **That** bag.

3. **あの** ボールペン.

  **That** ballpen **over there**.


###Examples: Other usage (with additional Noun)

1. **この** かさ は わたし の です.

  **This** bag is mine.

2. **その**  カメラ	も わたし の です.

  **That** camera is also mine.

3. **あの** ラジオ  は   わたし の です.

  **That** radio (over there) is mine.

##Part 3: **そうです** / **そうじゃありません**

In the case of a noun sentence, the word **そう** is often used to answer a question requiring an affirmative or negative  answer. **はい、そうです** is the affirmative  answer and **いいえ、そうじゃありません** is the negative answer.

###Examples: 

1. それは テレホンカード ですか.

  Is that a telephone card?

  Ans: *Positive form*

  はい、そうです

    … Yes, it is (lit. Yes, it’s so)

    

2. それは テレホンカード ですか.

  Is that a telephone card?

  Ans: *Negative form*

  いいえ、そうじゃありません.

    … No. it isn’t.

##Part 4: **S1 か**、**S2 か**

This is a question asking the listener to choose between alternatives, S1 and S2, for the answer to this type of question, the chosen sentence is stated.

Neither はい nor いいえ is used.

###Example:

1. これは **「９」ですか、「７」ですか**?

  Is this a "9" or a "7"?

  Ans: …「９」です.

  ... It's a "9".



##Part 5: **N1 の N2**

You learned in **Lesson 1** that **の** is used to connect two nouns when N1 modified N2. In **Lesson 2** you learned two other uses of this **の**.

###Examples: N1 explains what N2 is about.

1. これ は  コンピュータ **の** ほん です.

  This is a book **on** computer.
        This is "wa" computer "no" book "desu"

2. これ は くるま **の** ざっし です.

  This is a magazine **on** car.
        This is "wa" car "no" magazine "desu"

3. それ は  にほんご **の** ほん です.

  That is a book **on** Japanese.
        That is "wa" Japanese "no" book "desu"




###Examples: N1 explains who owns N2.

1. これは わたし**の**ほん です.

  This is my book.
        This is "wa" my "no" book "desu"

2. これ は Kevin さん **の** かばん です.

  This is Mr. Keven's bag.
        This is "wa" Mr.Kevin "no" bag "desu"





###Notes: だれ **の**  - Whose 

N2  is sometimes omitted when it is obvious.

When N2 means a person, however, you can not omit it.


###Examples:

1. あれは だれ **の** かばんですか

  Whose bag is that (over there)?
        That is (over there) "wa" whose bag "desu ka"

  Ans#1: Omitted N2

    Mikey さん **の** です.

    It is Mikey's.

  Ans#2: Omitted N2

    わたし **の** です

    It is mine.


2. この かばん は あなた **の** です か?

  Is this bag your's?
        This bag "wa" you "no" "desu ka"

  Ans#1: Omitted N2

  いいえ, わたし の じゃ ありません.

    No, it is not mine.

##Part 6: **そうですか** - "...Oh, I see"

This expression is used when the speaker receives  new information and shows that he or she understands it.

###Example: Conversation

Person2: …いいえ, ちがいます. Sasaki さんのです.

*Person2: ...No, you're wrong. It's Mr.Sasaki's*

##B: **こちら** / **そちら** / **あちら**

こちら, そちら and あちら are demonstratives refer to direction.

こちら, そちら and あちら are also used to refer to location, in which case, they are politer than ここ / そこ and / そこ.

##Part 1: N1 は  N2 (place) です

Using this sentence pattern, you can explain where a place, a thing or a person is.



###Examples:

1. おてあらい   は あそこです.

  The restroom is there.

2. でんわ  は ２かいです.

  The telephone is on the 2nd floor.

3. やまだ さん  は  じむしょ です.

  Mr. Yamada is in the office.


##Part 2: **どこ** / **どちら**

**どこ** means “where”.

**どちら** means “which direction”.

**どちら** can also mean “where”. in which case it’s politer than どこ.

###Examples:

1. おてあらい  は   **どこ** ですか?

  Where is the restroom?
        Restroom "wa" where "desu ka"

  Ans: あそこです。…

    It's there..

2. エレベーターは どちらですか

  Where is the elevator?
        Elevator "wa" where "desu ka"

  Ans: あちらです.

    It's in that direction (It's there)

###Notes:
**どこ** or **どちら** is also used to ask the name of a country, company, school or any place or organization a person belongs to. *You can not use なん (what)*.
**どちら** is politer than **どこ**.


###Examples:

1. がっこう は どこですか?

  What is the name of your school?
        School "wa" "doko/where" "desu ka"

2. かいしゃは どちらですか?

  What company do you work for?
        Company "wa" "dochira/where" "desu ka"

3. おくに は どちら ですか?

  Which country are you from?
        Country "wa" "dochira/where" "desu ka"

###Additional Notes: Another  **N1 の N2**

When **N1** is the **name of a country** and **N2** is a product, it means that N2 is made in that country.

When **N1** is the **name of a company** and **N2** is a product, it means that N2 is made by that company.

In this structure, **どこ** is used to ask where or by whom N2 is made.


###Example:

1. これ **は** どこ **の** コンピューター です か?

  Where is this computer made? Or Who is the maker of this computer?
          This "wa" where "no" computer "desu ka"

  Ans#1: にほん **の** コンピューター です.

  It’s made in Japan.

  Ans#2: Lenovo  **の** コンピューターです.

  Lenovo's computer.

##Part 3: **こ** / **そ** / **あ** / **ど** series

The こ / そ / あ / ど system of demonstrative words.

For things: **これ**,	 **それ**, 	**あれ**,	 **どれ** 

For things/person: **この**, 	**その**, 	**あの**, 	**どの N**

For place: **ここ**,	**そこ**,	**あそこ**,	**どこ**

Direction/place (polite): **こちら**,  **そちら**, 	**あちら**,  **どちら**

##Part 4: prefix **お**

The **prefix お** is added to a word concerning the listener or a third person in order to express the speaker’s respect to the person.

*Note*: 

Prefix  **お**  is use   for asking the listener only not use for one self !


###Examples:

1. **お** なまえ は なんです か?

  What is your name?
        "O" name "wa" what "desu ka"

2. **お** くに は どちら ですか?

  Which country do you from?
        "O" country "wa" where "desu ka"

3. お いくつ です か?

  How old are you?
        "O" how many "desu ka"

##Part 2: V **ます**

A verb with **ます** works as a predicate.
**ます** makes a sentence polite

**ます** is used when a sentence expresses a habitual thing or a truth. It is also used when a sentence expresses a thing that will occur in the future. The negative form and the forms in the past tense are shown below example.

**Affirmative Form:**

Present/Future: おき**ます**

Past: おき**ました**

**Negative form:**

Present/Future: おき**ません**

Past: おき**ませんでした**
